## Задания

### 1. Посчитать, сколько книг вышло после 1 января 2000 года;

```python
sql = '''SELECT COUNT(book_id)
         FROM books
         WHERE publication_date >= '2000-01-01' '''
select(sql)
```

Начиная с 1 января 2000 года опубликовалась 821 книга.

### Для каждой книги подсчёт количества обзоров и среднюю оценку;

```python
sql = '''SELECT rev.book_id, COUNT(DISTINCT rev.review_id) AS count_rev, AVG(rat.rating) AS avg_rating
         FROM reviews AS rev
         JOIN ratings rat on rev.book_id = rat.book_id         
         group by rev.book_id
         
         '''
select(sql)
```

Для каждой книги посчитали количество обзоров и средний рейтинг.

### Определение издательства, которое выпустило наибольшее число книг толще 50 страниц (с той целью, чтобы исключить из анализа брошюры).

```python
sql = '''SELECT pub.publisher, COUNT(DISTINCT b.book_id) AS count_book
         FROM publishers pub
         JOIN books b ON pub.publisher_id = b.publisher_id
         WHERE b.num_pages > 50
         GROUP BY pub.publisher
         ORDER BY count_book DESC
         LIMIT 1 '''
select(sql)
```

В ходе работы с БД выяснили, что издательство "Penguin Books" выпустило наибольшее число книгу с толщиной более 50 страниц.

###  Определение автора с самой высокой средней оценкой книг — учитывать только книги с 50 и более оценками.

```python
sql = ''' select t.author, max(t.rat_avg) as rat_max from
(select s.author as author,avg(r.rating) as rat_avg, count(r.rating_id) as rat_cnt from
(select * from authors a
inner join books b on b.author_id = a.author_id) s
inner join ratings r on r.book_id = s.book_id
group by s.author) t
where t.rat_cnt > 50
group by t.author
order by rat_max desc
limit 1
'''
select(sql)
```

##  Подсчёт среднего количества обзоров от пользователей, которые поставили больше 50 оценок.

```python
sql = '''SELECT AVG(s.review_count) FROM 
(SELECT r.username as username, COUNT(DISTINCT r.review_id) as review_count 
FROM reviews r
JOIN (SELECT t.username as username, COUNT(t.rating_id) as rating_count FROM ratings t
GROUP BY t.username
HAVING count(t.rating_id) > 50) s on s.username = r.username
GROUP BY r.username) s
'''
select(sql)
```

В среднем, 24 пользователя сделали обзор, при этом поставив более 50 оценок.
